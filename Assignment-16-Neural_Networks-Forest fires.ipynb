# Assignment-16-Neural_Networks - Forest fires

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
```

```python
forest = pd.read_csv('forestfires.csv')
forest
```

```python
forest.info()
```

# Applying PCA due to having more no. of columns

```python
# Scaling the data (Leaving out the target variable, and taking only numerical data for input)
df1= forest.iloc[:,2:30 ]

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

sc.fit(df1)
forest_norm = sc.transform(df1)
forest_norm

# Normalised dataset
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 28)
pca_values = pca.fit_transform(forest_norm)
pca_values
```

```python
# The amount of variance that each PCA explains is 

var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance

var1 = np.cumsum(np.round(var,decimals = 4)*100)

var1
```

```python
# Variance plot for PCA components obtained 

plt.figure(figsize=(12,4))
plt.plot(var1, color='green');
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:24],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7',
                                                             'pc8','pc9','pc10','pc11','pc12','pc13','pc14',
                                                             'pc15','pc16','pc17','pc18','pc19','pc20','pc21',
                                                             'pc22','pc23','pc24']),
                     forest[['size_category']]], axis = 1)
finalDf.size_category.replace(('large','small'),(1,0),inplace=True)
finalDf                               
```

```python
# split into input (X) and output (Y) variables

array = finalDf.values
X = array[:,0:24]
Y = array[:,24]

X.reshape(1,-1)
Y.reshape(-1,1)
```

```python
# create model

model = Sequential()
model.add(Dense(12, input_dim=24, activation='relu'))
model.add(Dense(8, activation ='relu'))
model.add(Dense(1, activation ='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer= 'adam', metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(X,Y, validation_split=0.3, epochs=150, batch_size= 10)
```

```python
# Evaluate the model
scores= model.evaluate(X, Y)
print('%s: %.2f%%' % (model.metrics_names[1], scores[1]*100))
```

```python
history.history.keys()
```

```python
# Summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('accuracy')
plt.ylabel('epoch')
plt.legend(['train', 'test'])
plt.show()

#summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.xlabel('loss')
plt.ylabel('epoch')
plt.legend(['train', 'test'])
plt.show
```
